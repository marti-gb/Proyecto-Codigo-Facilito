```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from pandas.plotting import scatter_matrix

dfPNatalidad = pd.read_csv('/Users/User/Desktop/Proyecto-CF/proyecto/data/Consolidado/Datos-Proyeccion-Natalidad.csv')
dfPNatalidad
```

```python
dfDatos = dfPNatalidad[['FPI (Nominal)', 'Precio Brent', 'Mundo (Inflacion)', 'Mundo (Natalidad)']]
```

```python
corr_matrix = dfDatos.corr(method='pearson')
corr_matrix
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))

sns.heatmap(
    corr_matrix, annot = True, cbar = False, annot_kws = {"size": 8}, vmin = -1,
    vmax = 1,center = 0, cmap = sns.diverging_palette(20, 220, n=200),
    square = True, ax = ax
)

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation = 45,
    horizontalalignment = 'right',
)

ax.tick_params(labelsize = 10)
```

```python
scatter_matrix(dfDatos)
plt.show()
```

```python
plt.plot(dfPNatalidad['Año'], dfPNatalidad['FPI (Nominal)'], marker='o', linestyle='-', color='b')
plt.title('Tendencia de Datos Históricos del FPI (Nominal)')
plt.xlabel('Años')
plt.ylabel('Valores')
plt.grid(True)

coefficients = np.polyfit(dfPNatalidad['Año'], dfPNatalidad['FPI (Nominal)'], 1)
trend_line = np.polyval(coefficients, dfPNatalidad['Año'])
plt.plot(dfPNatalidad['Año'], trend_line, label='Línea de Tendencia', linestyle='--', color='red')
plt.legend()
plt.show()
```
