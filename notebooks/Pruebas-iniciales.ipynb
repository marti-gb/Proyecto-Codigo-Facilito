```python
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

#Cargar los datos en un data frame
data = pd.read_csv('/Users/User/Desktop/Proyecto-CF/proyecto/data/Proyecto-Natalidad.csv')  # Reemplaza 'datos.csv' con el nombre de tu archivo de datos

# Dividir los datos en caractersticas (X) y etiquetas (y)
X = data.iloc[:, 0:3]  # Primeras tres columnas como caractersticas
y = data.iloc[:, 3]    # ltima columna como etiqueta
#print(X.to_numpy().shape[1:])
# print(y)

#scaler = MinMaxScaler()
#X_scaled = scaler.fit_transform(X)


# Dividir los datos en conjuntos de entrenamiento y prueba
#X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, shuffle=False)
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
X_train_full, X_test, y_train_full, y_test = train_test_split( X, y,test_size=0.2, shuffle=False)
X_train, X_valid, y_train, y_valid = train_test_split( X_train_full, y_train_full,test_size=0.2,  shuffle=False)
#print(X_train_full)
#print(X_train)

norm_layer = tf.keras.layers.Normalization(input_shape=X_train.shape[1:])

#Crear el modelo de red neuronal
model = tf.keras.Sequential([
    norm_layer,
    # tf.keras.layers.Dense(64, activation= 'relu', input_shape=(3,)),
    # tf.keras.layers.Dense(96, activation= tf.keras.activations.selu),
    tf.keras.layers.Dense(300, input_dim=2, activation=tf.keras.layers.ELU(0.3)),
     tf.keras.layers.Dense(300, input_dim=2, activation=tf.keras.layers.ELU(0.3)),
    tf.keras.layers.Dense(300, activation=tf.keras.layers.ELU(0.3)),
    tf.keras.layers.Dense(300, activation=tf.keras.layers.ELU(0.3)),
    tf.keras.layers.Dense(300, activation=tf.keras.layers.ELU(0.3)),
    # tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compilar el modelo
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
norm_layer.adapt(X_train)

# Definir la funcin para guardar el historial de entrenamiento
class CustomCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        print(f"Epoch {epoch+1}, Loss: {logs['loss']:.4f}, Val Loss: {logs['val_loss']:.4f}")

# Entrenar el modelo
history = model.fit(X_train, y_train, epochs=2500, batch_size=32, validation_data=(X_test, y_test)
                   # , callbacks=[CustomCallback()]
                   )

loss_eval = model.evaluate(X_test, y_test)
print('Prdida en los datos de prueba: ', loss_eval)
#Graficar el historial de entrenamiento
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'b', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Realizar predicciones en nuevos datos
new_data = pd.DataFrame([[2022, 100.93, 7.96757361617885], [2023, 82.49, 7]])  # Ejemplo de nuevos datos
#new_data = pd.DataFrame([[100.93, 7.96757361617885], [82.49, 7]])
predictions = model.predict(new_data)

# Imprimir las predicciones
print(predictions)
```
